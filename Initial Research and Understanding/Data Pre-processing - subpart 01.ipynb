#### <i>AGAIN EXTRACTING REQUIRED COLUMNS</i>

```python
data_extracted_2 = cbc.iloc[:, 53:-1]
```

#### <i>REPLACING OTHER TYPES OF MISSING DATA WITH NAN</i>

```python
unknown_values = ['ERROR', '----', '++++']
data_extracted_2 = data_extracted_2.replace(unknown_values,np.NaN)
data_extracted_2
```

#### <i>AGAIN CHECKING FOR NULL VALUES</i>

```python
data_extracted_2.isna().sum()
```

#### <i> FILLING NAN WITH MEAN VALUES OF THEIR COLUMNS</i>

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean', missing_values=np.nan)
imputer = imputer.fit(data_extracted_2)
data_extracted_2.iloc[:,:] = imputer.transform(data_extracted_2)

data_extracted_2
```
